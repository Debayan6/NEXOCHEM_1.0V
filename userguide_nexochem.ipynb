```python
#importing necessary modules
from nexochem import *
import numpy as np
```

## WASP-12b

```python
# loading Inputs                                                                                  

p_t_data = np.loadtxt('../input/wasp12.dat', skiprows=2)

temperature = p_t_data[:,0] 
pressure = p_t_data[:,1]
```

```python
# Easy plot

plot_species = ['C1O2', 'C1O1', 'C1H4','H3N1','H2S1']
nexochem_plot(plot_species, temperature, pressure)
```

```python
# Advanced plot 
# changing metallicity, C/O ratio
plot_species = ['C1O2', 'C1O1', 'C1H4','H3N1','H2S1']
nexochem_plot(plot_species, temperature, pressure,meta= 1,c_o= 1.5)
```
